```sage
# Une parabole classique : x² - 4
parabole = plot(x^2 - 4, (x, -5, 5), color='blue', thickness=2, legend_label="Polynôme degré 2 normal")
# La ligne des racines (y=0)
axe_zero = plot(0, (x, -5, 5), color='red', linestyle='--')

show(parabole + axe_zero, title="Un degré 2 a au MAX 2 racines")
```

```sage
# Les 3 points par lesquels on DOIT passer (tes racines a, b, c)
points_racines = point([(-3,0), (0,0), (3,0)], color='green', size=50, zorder=5, legend_label="Racines a, b, c")

# Une tentative de parabole : elle en rate forcément au moins un !
parabole_ratee = plot(x^2 - 3, (x, -5, 5), color='blue', linestyle=':')

show(points_racines + parabole_ratee + axe_zero, title="Impossible pour une parabole de toucher les 3 !")
```
