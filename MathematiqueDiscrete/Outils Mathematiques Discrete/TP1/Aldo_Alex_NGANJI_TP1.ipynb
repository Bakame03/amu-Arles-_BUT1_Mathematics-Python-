3. Quelles sont les expressions n’admettant aucune factorisation ? Conjecturez un résultat général.

Aucune factorisation quand n = 1, 2, 4, 8, 16 les puissances de 2
Resultat general:

```python
x = var('x')
y = var('y')

def est_puissance_de_deux(n):
    while n % 2 == 0:  # Tant que n est pair
        n = n / 2     # On le divise par 2
    return n == 1      # S'il reste 1, c'est gagné. Sinon, il y a un facteur impair.

for n in range(1,18):
    if est_puissance_de_deux(n) == True:
        print(f"pas de factorisation pour n={n}")
    else:
        print(factor(x^n + y^n))
```

3 Résolution d’équation

```python
Nous abordons maintenant les équations et leur résolution. Les fonctions principales utilisées sont solve() et
find_root() .
Considérons l’équation x^2 − 2 = 0 qui admet deux solutions égales à √2 et −√2.
```

```python
x = var('x')
eq = x**2 - 2
solutions = solve(eq, x) # On précise la variable inconnue
print(solutions)
```

```python
L’ensemble des solutions est retourné sous forme de liste.
```

```python
solutions[0]
```

```python
solutions[1]
```

Pour effectuer des résolutions numériques d’équations (c’est-à-dire donnant une valeur approchée d’une solution), on
utilise la fonction find_root() qui prend en argument une fonction d’une variable ou une égalité symbolique, et les
bornes de l’intervalle dans lequel il faut chercher une solution.

```python
expr = x**2-2
sol = find_root(expr, -2, 2)
sol
```

Quelles différence remarquez vous ?
find_root retourne un resultat whereas solve returns 2 resultats

Il est important de préciser la variable représentant l’inconnue, en particulier si l’équation admet un paramètre. Par
exemple, si l’on veut résoudre l’équation
x^2 − ((2 / cos φ)*x) + (5/(cos^2 φ)) − 4 = 0 

avec φ ∈ ]− π/2 ; π/2[

```python
x, phi = var('x, phi')
eq = x**2 - 2/cos(phi)*x + 5/cos(phi)**2 - 4 == 0
eq
sol = solve(eq,x)
sol
```

Exercice 4. On considère trois réels distincts a, b, c et on définit le polynôme :
voir le polynôme dans le pdf

1. Définir trois variables symboliques par a, b, c = var('a', 'b', 'c') et créer une fonction
polynome(x) qui renvoie P (x).

```python
a, b, c = var('a, b, c')
def polynome(x):
    eq = ((x-a)*(x-b))/((c-a)*(c-b))+((x-b)*(x-c))/((a-b)*(a-c))+((x-a)*(x-c))/((b-a)*(b-c))-1
    return eq
print(polynome(5))
```

2. Calculer P (a), P (b) det P (c).

```python
print(polynome(a))
print(polynome(b))
print(polynome(c))
```

3. Comment un polynôme de degré 2 pourrait avoir 3 racines distinctes ? Expliquez ! (Vous pourrez appliquer la
méthode full_simplify() à votre fonction ; n’hésitez pas à demander de l’aide et/ou des explications à votre
charmant professeur !)

Un polynôme de degré 2 ne peut jamais avoir trois racines distinctes (il en a au maximum deux). Le fait de trouver P(a), P(b) et P(c) tous égaux à zéro montre que ce n'est pas un polynôme de degré 2 classique. Puisqu'on a trois valeurs qui l'annulent, cela veut dire que ce polynôme est nul (plat), donc toujours égal à zéro. On le voit bien d'ailleurs si on le simplifie avec la méthode simplify_rational()

```python
p(x) = ((x-a)*(x-b))/((c-a)*(c-b))+((x-b)*(x-c))/((a-b)*(a-c))+((x-a)*(x-c))/((b-a)*(b-c))-1


P_simplifie = p(x).simplify_rational()

print(f"L'expression simplifiée est : {P_simplifie}")
```

Exercice 5. (Le nombre d’or)

```python
phi = var('phi')
equation = phi^2 - phi - 1 == 0
solutions = solve(equation, phi)
print(solutions)

nombredor = solutions[1].rhs()
print(nombredor.n())
```
