Exemple 1. On souhaite repr√©senter les premiers termes de la suite num√©rique :
voir la forme dans le pdf
Le code ci-dessous permet de repr√©senter les 20 premiers termes (de u0 jusqu‚Äô√† u19).

```python
# Formule de r√©currence
def suite(n) :
    u0 = 0.5
    U = [u0]
    for i in range(1,n) :
        u = (sin(U[i-1]) / (i*U[i-1])).n()
        U.append(u)
    return U
# Liste des coordonn√©es des points
n = 20
U = suite(n)
Coord = [(i, U[i]) for i in range(len(U))]
## affichage du graphique
G = line(Coord) + point(Coord)
G.show()
```
