# Chiffrement par blocs

##### Le chiffrement que nous allons étudier a été publié par Lester S.Hillen 1929. C’est un chiffrement polygraphique, c’est-à-dire qu’on ne(dé)chiffre pas les lettres les unes après les autres, mais par paquets en utilisant des matrices à coefficients dans Z/26Z.

## 1 Méthode de chiffrement

#### Pour coder un message selon ce procédé, on suit les étapes :
#### ➣ Étape 1 : On regroupe les lettres du message deux par deux ;
#### ➣ Étape 2 : On remplace chaque lettre par un nombre, comme indiqué par le tableau suivant :
#### Lettre(Nombre) : a(0) b(1) c(2) d(3) e(4) f(5) g(6) h(7) i(8) j(9) k(10) l(11) m(12) n(13) o(14) p(15) q(16) r(17) s(18) t(19) u(20) v(21) w(22) x(23) y(24) z(25)

#### Exemple 1. On veut par exemple coder le message « j’adore les maths ».
#### On le décompose en ja-do-re-le-sm-at-hs puis on remplace par :
#### (9; 0) − (3; 14) − (17; 4) − (11; 4) − (18; 12) − (0; 19) − (7; 18)
#### Remarque. Si le nombre de lettres du message avait été impair, on aurait ajouté une lettre arbitraire à la fin.

#### ➣ Étape 3 : Chaque couple de nombres (x; y) de la liste précédente est transformé en un nouveau couple (x′; y′)
#### de nombres entiers compris entre 0 et 25, à l’aide d’une matrice A = [[a, b],[c, d]]
#### via la relation
#### A(x y) = (x′ y′)
#### c’est à dire
#### ax + by ≡ x′ (mod 26)
#### cx + dy ≡ y′ (mod 26)
#### Vocabulaire. La matrice A est appelé la clé du chiffrement.
#### ➣ Étape 4 : Ces deux nombres x′ et y′ sont transformés en lettres en utilisant le tableau de correspondance.

# 3 Chiffrement

##### Écrire avec sagemath une fonction chiffre_hill(mess_clair) qui prend en entrée la chaîne de caractère
##### mess_clair et qui retourne le message chiffré. Ne pas oublier d’enlever les espaces dans la chaîne de caractère.

```sage
mess_clair = "Je ne suis pas un numero je suis je suis un homme libre"

def lettreToEntier(lettre, alphabet = "abcdefghijklmnopqrstuvwxyz"):
    return alphabet.find(lettre)

def entierToLettre(a, alphabet = "abcdefghijklmnopqrstuvwxyz"):
    return alphabet[a]

def chiffre_hill(mess_clair):
    key_matrix = matrix([[9, 4], [5, 7]])
    mess_clair = mess_clair.replace(" ", "")
    if len(mess_clair) % 2 == 1:
        mess_clair += 'a'  
    mess_list = []
    mess_chiffre = ""
    for i in range(0, len(mess_clair), 2):
        nb1 = mess_clair[i]
        nb2 = mess_clair[i + 1] 
        mess_list.append([lettreToEntier(nb1), lettreToEntier(nb2)])
    for pair in mess_list:
        vecteur = matrix([[pair[0]], [pair[1]]])
        produit = key_matrix * vecteur
        produit = produit % 26
        mess_chiffre += entierToLettre(int(produit[0][0])) + entierToLettre(int(produit[1][0]))
    return mess_chiffre

print("test d'un message chiffré avec Hill : ", chiffre_hill(mess_clair))

```
