## Exercice 1. Écrire une fonction minimum qui trouve la valeur minimale des coefficients d’une matrice A et sa position
## (i_min,j_min) dans la matrice A.

```python
def minimum(A) :
    n = A.nrows()
    p = A.ncols()
    min = A[0,0]
    istValeurMinPlusPosition = []
    for i in range(n) :
        for j in range(p) :
            if A[i,j] < min :
                min = A[i,j]
                listValeurMinPlusPosition = [min, i, j]
    return listValeurMinPlusPosition
A = matrix([[2, -1, -1, -2], [4, 4, 2, 0], [0, 4, -2, 0], [-1, -1, 0, -3]])
print("la valeur minimale est ", minimum(A)[0], "et sa position est (", minimum(A)[1], ",", minimum(A)[2], ")")
```

### Remarque importante
### Si on pose A = matrix([[1,2],[3,4]]) , puis B = A , puis que l’on modife la matrice A par A[0,0] = 7 . Que se
### passe-t-il ?

```python
A=matrix([[1,2],[3,4]])
B = A
A[0, 0] = 7
A, B
```

### La matrice B est aussi modifée !

#### Explications
#### — A ne contient pas les coefficients de la matrice, mais l’adresse où sont stockés ces coefficients (c’est plus léger
#### de manipuler l’adresse d’une matrice que toute la matrice).
#### — Comme A et B pointent vers la meme zone qui a été modifée, les deux matrices A et B sont modifées.
#### — Pour pouvoir défnir une copie de A , avec une nouvelle adresse mais les memes coefficients, on écrit 
##  B = copy(A) 
#### Les modifcations sur A ne changeront plus B.

##  Exercice 2. La méthode du pivot de Gauss repose sur trois opérations élémentaires les lignes :
##  — opération 1 : Li <- cLi avec c != 0 : on multiplie une ligne par un réel ;
##  — opération 2 : Li <- Li + cLj : on ajoute à la ligne Li un multiple d’une autre ligne Lj ;
##  — opération 3 : Li <-> Lj : on échange deux lignes.

### Écrire les fonctions op1(A, i, c) , op2(A, i, j, c) et op3(A, i, j) correspondant à ces trois opérations élémentaires.
### Attention !! la matrice A ne devra pas etre modifée pendant l’exécution de cette fonction ! Pour cela, il faudra écrire
### AA = copy(A) ou AA = A.copy() et travailler sur cette copie.

```python
A=matrix([[1,2],[3,4],[5,6],[7,8]])

print(A)
print("---")

def op1(A, i, c):
    if(c==0):
        print("Erreur: le coefficient c ne doit pas être zéro.")
        return A
    AA = copy(A)
    n = A.ncols()
    for j in range(n):
        AA[i, j] = A[i, j] * c
    return AA

print(op1(A, 2, 3))
print("---")

def op2(A, i, j, c):
    if(c==0):
        print("Erreur: le coefficient c ne doit pas être zéro.")
        return A
    AA = copy(A)
    n = A.ncols()
    for k in range(n):
        AA[i, k] = A[i, k] + c * A[j, k]
    return AA

print(op2(A, 2, 1, 2))
print("---")

def op3(A, i, j):
    AA = copy(A)
    n = A.ncols()
    for k in range(n):
        temp = AA[i, k]
        AA[i, k] = AA[j, k]
        AA[j, k] = temp
    return AA

print(op3(A, 0, 2))
print("---")
```
