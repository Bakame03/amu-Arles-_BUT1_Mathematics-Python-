```python
def parties_aliquotes(x):
    diviseurs = []
    for i in range(1,(x)):
        if x%i == 0:
            diviseurs.append(i)
    return diviseurs

def suivant_aliquote(x):
    getDiviseursX = parties_aliquotes(x)
    next_aliquote = 0
    for truc in getDiviseursX:
        next_aliquote += truc
    return next_aliquote

def amicaux(x):
    amicaux = []
    for value in range(1,x):
        if suivant_aliquote(suivant_aliquote(value)) == value and value != suivant_aliquote(value):
            amicaux.append(value)
    return amicaux

L = amicaux(10000)
print(L)
```
