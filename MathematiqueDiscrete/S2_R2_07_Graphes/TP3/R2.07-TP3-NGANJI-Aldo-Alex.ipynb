# 2 Algorithme de Bellman

#### L’algorithme de Bellamn permet de déterminer les plus courts chemins entre le sommet no 1 et les autres sommets
#### pour un graphe acyclique dont le sommets sont triés dans l’ordre topologique.
#### Remarque. pour trier les sommets d’un graphe g dans l’ordre topologique, on peut utiliser la méthode topological_sort()

#### Remarques. 
#### 1. sagemath possède une représentation de l’infini : Infinity .
#### 2. Infinity vérifie les propriétés classiques :
