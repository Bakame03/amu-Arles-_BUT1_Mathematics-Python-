```python
def Bellman_Ford(M):
    nbSom = M.nrows()                     
    distances = [Infinity] * nbSom             
    distances[0] = 0                       

    change = True                 
    while change:                 
        change = False            
        for i in range(nbSom):            
            for j in range(nbSom):        
                pd = M[i][j]
                if pd == 0 and i != j:
                    pd = Infinity  
                if distances[i] + pd < distances[j]:
                    distances[j] = distances[i] + pd
                    change = True 
    return distances

print(Bellman_Ford(M))
print(Bellman_Ford(M2))
```

```python
def dijkstra(M):
    nbrS = M.nrows()                     
    distances = [Infinity] * nbrS             
    distances[0] = 0                       
    pointeur = [0] * nbrS                

    while 0 in pointeur:              
        petit = Infinity
        
        i = -1
        for m in range(nbrS):
            if pointeur[m] == 0 and distances[m] < petit:
                petit = distances[m]
                i = m

        if i == -1:                   
            break

        pointeur[i] = 1               

        for j in range(nbrS):            
            if pointeur[j] == 0:
                poids = M[i][j]
                if poids == 0 and i != j:
                    poids = Infinity  
                if distances[i] + poids < distances[j]:
                    distances[j] = distances[i] + poids
    return distances

print(dijkstra(M))
print(dijkstra(M2))
```
