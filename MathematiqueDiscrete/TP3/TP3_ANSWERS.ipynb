```python
def IsPrime(n):
    sum = 0
    for i in range(1, n+1):
        if n%i == 0:
            sum += 1
    if sum == 2:
        return True
    return False

def premiers(n):
    listePremiers = []
    for i in range(2, n+1):
        if IsPrime(i):
            listePremiers.append(i)
    return listePremiers
L1 = premiers(100)
print(L1)
    
```
