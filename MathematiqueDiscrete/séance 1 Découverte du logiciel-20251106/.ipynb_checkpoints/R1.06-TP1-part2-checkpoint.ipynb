## Tests, boucles

Voici deux programmes écrits en `Sagemath` :

**Programme 1**

```
n = int(input('valeur de n ? '))
epaisseur = 0.2
for i in range(n) :
    epaisseur = epaisseur*2  # autre syntaxe epaisseur *= 2
print(epaisseur)
```

**Programme 2**

```
h = float(input('Quelle hauteur ? '))
n = 0
epaisseur = 0.2
while epaisseur < h :
    epaisseur = epaisseur*2  # autre syntaxe epaisseur *= 2
    n = n+1  # autre syntaxe n += 1
print(n)
```

Ces deux programmes servent à résoudre le problème suivant :
